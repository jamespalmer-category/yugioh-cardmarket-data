```python
#want to enter the above df and then get back the df with a column full of set codes.

def set_code(dataframe):
    
    
    code_list = []
    for i in range(0,len(dataframe)):
        url = dataframe['Yugipedia url'][i]
        print(url)
        result = requests.get(url)
        src = result.content
        soup = BeautifulSoup(src, 'html.parser')
        
        card_info_headers = soup.find_all(attrs = {'class':'set-list'})
        
        if card_info_headers[0].find('td').text[0:4] == '25TH':
            code = 'BACH'
        elif card_info_headers[0].find('td').text[0:4] == '\"Blu':
            code = 'DUOV'
        else:
            code = card_info_headers[0].find('td').text
            code = code[0:4]
            
        if code[3] == '-':
            code = code[0:3]
            
        print(code)
        code_list.append(code)
    
    dataframe['Set Code'] = code_list
    return dataframe
```
