```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re
import os
import cardmarket_singles_parsing as csm
from TCG_set_V4 import TCG_set
```

```python
url = "https://yugipedia.com/wiki/Order_of_Set_Release#2022_2"
result = requests.get(url)
src = result.content
soup = BeautifulSoup(src, 'html.parser')

TCG_set_name = []
TCG_set_yugipedia_link = []

for TCG_card_set in soup.find_all("ul")[339:576]:
    set_names = TCG_card_set.find_all('a')
    for set_name in set_names:
        TCG_set_name.append(set_name.text)
        TCG_set_yugipedia_link.append('https://yugipedia.com' + set_name.attrs['href'])
        
df = pd.DataFrame({"Set Name" : TCG_set_name, "Yugipedia url" : TCG_set_yugipedia_link})

df = df.drop_duplicates()
df = df.reset_index()
df = df.drop(columns = 'index')

sets_done_index = []

#Cards not on sale removed, special edition removed as Cardmarket covers them anyway.
for i in range(0, len(df)):
    if ('KC Grand Tournament' in list(df['Set Name'])[i]) or ('Japan Championship' in list(df['Set Name'])[i]) or ('Masterpiece Series' in list(df['Set Name'])[i]) or ('Premiere! promotional card' in list(df['Set Name'])[i]) or ('Sneak Peek Participation Card' in list(df['Set Name'])[i]) or ('Special Edition' in list(df['Set Name'])[i]):
        sets_done_index.append(i)


df = df.drop(sets_done_index)
df = df.iloc[::-1]
df = df.reset_index()

#All LART but one removed
lost_art_index = []
for i in range(37, len(df)):
    if 'The Lost Art Promotion' in list(df['Set Name'])[i]:
        
        lost_art_index.append(i)

df = df.drop(lost_art_index)
df = df.drop(columns = 'index')
df = df.reset_index()
df = df.drop(columns = 'index')
```
